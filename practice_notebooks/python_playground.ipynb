```python
nums = [1,2,3,4]
def product_except_self(nums):
    if not nums:
        return []

    left = [1]
    right = [1]
    answer = []


    for i in range(1,len(nums)):
        left.append(left[-1] * nums[i - 1])
    
    for i in range(len(nums)-2, -1, -1):
        right.append(right[-1] * nums[i + 1])

    right.reverse()

    for i in range(len(nums)):
        answer.append(left[i] * right[i])
    
    return answer

product_except_self(nums)
```

```python
nums = [100,4,200,1,3,2]
def longest_cons_seq(nums):
    seen = set(nums)

    current = 0
    length = 0

    for num in seen:
        if (num - 1) not in seen:
            current = num
            length = 1
            
            while current + 1 in seen:
                current += 1
                length += 1
            
            print(num,length)

longest_cons_seq(nums)
```

```python
nums = [1,1,1,2,2,3]
k = 2

def top_k_freq_element(nums,k):
    if not nums:
        return 0
    
    freq = {}
    result = []
    
    for num in nums:
        freq[num] = freq.get(num,0) + 1

    sorted_items = sorted(freq.items(),key=lambda x:x[1],reverse=True)

    print(sorted_items)

    for i in range(0,k):
        result.append(sorted_items[i][0])
    
    return result

top_k_freq_element(nums,k)
```
